```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
import random
import pandas as pd

# === Load fitur dari .pkl ===
with open("features_cloudinary.pkl", "rb") as f:
    features, filenames = pickle.load(f)

# === Sampling 10 gambar secara acak ===
subset_indices = random.sample(range(len(features)), 10)
subset_features = features[subset_indices]
subset_filenames = [filenames[i] for i in subset_indices]

# === Hitung cosine similarity matrix ===
similarity_matrix = cosine_similarity(subset_features)

# === Buat heatmap dengan seaborn ===
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix,
            xticklabels=subset_filenames,
            yticklabels=subset_filenames,
            cmap="YlGnBu",
            annot=True,
            fmt=".2f")
plt.title("Cosine Similarity Heatmap (10 Gambar)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
